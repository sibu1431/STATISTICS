Q3.write a python function to calculate the probability distribution function (PDF) of a bernoulli distribution?

```python
def bernoulli_pmf(x, p):
    """
    Calculate the PMF of a Bernoulli distribution.

    Parameters:
        x (int): The value (0 or 1)
        p (float): The probability of success (0 â‰¤ p â‰¤ 1)

    Returns:
        float: The probability of observing value x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1")

    return p**x * (1 - p)**(1 - x)

# Example usage:
print("P(X=0):", bernoulli_pmf(0, 0.7))
print("P(X=1):", bernoulli_pmf(1, 0.7))
```

Q6.write a python program to calculate and plot cumulative  distribution function (cdf) of a discrete uniform distribution?

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for discrete uniform distribution
a = 1   # Minimum value
b = 6   # Maximum value (inclusive)

# Values in the support
x = np.arange(a, b + 1)

# Probabilities for each value in discrete uniform distribution
pmf = np.full_like(x, 1 / len(x), dtype=float)

# Calculate cumulative distribution function (CDF)
cdf = np.cumsum(pmf)

# Plot the CDF
plt.step(x, cdf, where='mid', label='CDF', color='blue', linewidth=2)
plt.title('CDF of Discrete Uniform Distribution')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.xticks(x)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

Q9.write a python function to calcLCULATE z-SCORE FROM A data set and plot them?



```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate z-scores from a dataset and plot them.

    Parameters:
    data (array-like): Input dataset (list or numpy array)
    """
    # Convert data to a NumPy array
    data = np.array(data)

    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data)

    # Compute z-scores
    z_scores = (data - mean) / std

    # Plot histogram of z-scores
    plt.hist(z_scores, bins=30, edgecolor='black', alpha=0.7, density=True)
    plt.title('Histogram of Z-Scores')
    plt.xlabel('Z-Score')
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    return z_scores

# Example usage with simulated normal data
```

Q15.implement hypothesis testing using Z-statistics for a sample dataset

```python
import numpy as np
from scipy.stats import norm

# Sample dataset
sample_data = [52, 55, 60, 58, 62, 57, 59, 61, 63, 60]

# Known population parameters
population_mean = 60   # Î¼â‚€ (hypothesized mean)
population_std = 5     # Ïƒ (known standard deviation)
n = len(sample_data)

# Calculate sample mean
sample_mean = np.mean(sample_data)

# Calculate Z-statistic
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Calculate two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Set significance level
alpha = 0.05

# Output results
print("Sample Mean:", sample_mean)
print("Z-Statistic:", z_score)
print("P-Value:", p_value)

# Decision
if p_value < alpha:
    print("Reject the null hypothesis (Hâ‚€).")
else:
    print("Fail to reject the null hypothesis (Hâ‚€).")
```

Q16.create a confidence interval for a dataset using python and interpret the result.

```python
import numpy as np
import scipy.stats as stats

# Sample dataset
data = [52, 55, 60, 58, 62, 57, 59, 61, 63, 60]

# Sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # use ddof=1 for sample standard deviation
n = len(data)

# Confidence level
confidence = 0.95
alpha = 1 - confidence

# Critical value (two-tailed t-distribution)
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

# Margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(n))

# Confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Output
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Confidence Interval ({int(confidence*100)}%): ({ci_lower:.2f}, {ci_upper:.2f})")
```

Q1.write a python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Calculate standard error
    standard_error = population_std / np.sqrt(n)

    # Compute Z-score
    z_score = (sample_mean - population_mean) / standard_error

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Interpretation
    alpha = 0.05
    if p_value < alpha:
        result = "Reject the null hypothesis: The sample mean is significantly different from the population mean."
    else:
        result = "Fail to reject the null hypothesis: No significant difference between the sample and population mean."

    return {
        'Sample Mean': sample_mean,
        'Z-score': z_score,
        'P-value': p_value,
        'Conclusion': result
    }

# Example usage
sample = [105, 110, 98, 100, 102, 108, 95, 103, 106, 104]  # Sample data
population_mean = 100
population_std = 10

result = z_test(sample, population_mean, population_std)
for key, value in result.items():
    print(f"{key}: {value}")
```

Q5.create a python function that calculate and visualize type1 andtype-2 error during  hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(mu0, mu1, sigma, n, alpha=0.05):
    """
    Visualize Type I and Type II errors for hypothesis testing.

    Parameters:
    - mu0: Mean under H0
    - mu1: Mean under H1
    - sigma: Standard deviation (assumed known)
    - n: Sample size
    - alpha: Significance level (default = 0.05)
    """

    se = sigma / np.sqrt(n)

    # Critical z value for two-tailed test
    z_critical = norm.ppf(1 - alpha / 2)

    # Critical sample mean boundaries
    x_crit_low = mu0 - z_critical * se
    x_crit_high = mu0 + z_critical * se

    # Type II error (Î²) under H1
    beta = norm.cdf(x_crit_high, loc=mu1, scale=se) - norm.cdf(x_crit_low, loc=mu1, scale=se)

    # X-axis range for plotting
    x = np.linspace(mu0 - 4 * se, mu1 + 4 * se, 1000)

    # PDFs for H0 and H1
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label=f"Hâ‚€: Î¼ = {mu0}", color='blue')
    plt.plot(x, y1, label=f"Hâ‚: Î¼ = {mu1}", color='green')

    # Shade Type I Error regions under H0
    plt.fill_between(x, y0, where=(x < x_crit_low) | (x > x_crit_high), color='red', alpha=0.3, label='Type I Error (Î±)')

    # Shade Type II Error region under H1
    plt.fill_between(x, y1, where=(x >= x_crit_low) & (x <= x_crit_high), color='orange', alpha=0.4, label='Type II Error (Î²)')

    # Critical lines
    plt.axvline(x_crit_low, color='black', linestyle='--', label='Critical Values')
    plt.axvline(x_crit_high, color='black', linestyle='--')

    # Labels
    plt.title("Type I and Type II Error Visualization")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Output
    print(f"Critical Region: [{x_crit_low:.2f}, {x_crit_high:.2f}]")
    print(f"Type II Error (Î²): {beta:.4f}")
    print(f"Power of the test (1 - Î²): {1 - beta:.4f}")

# Example usage:
visualize_type1_type2(mu0=100, mu1=105, sigma=15, n=30, alpha=0.05)
```

Q9. write a python function to calculate the confidence interval for a sample mean and explains its significance.

```python
import numpy as np
from scipy import stats

def confidence_interval(sample_data, confidence=0.95):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # sample standard deviation (unbiased)

    # Calculate the critical t-value
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n - 1)

    # Standard error of the mean
    sem = sample_std / np.sqrt(n)

    # Margin of error
    margin_of_error = t_critical * sem

    # Confidence interval
    ci_lower = sample_mean - margin_of_error
    ci_upper = sample_mean + margin_of_error

    return (ci_lower, ci_upper), sample_mean, margin_of_error

# Example usage
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=30)
ci, mean, moe = confidence_interval(data, confidence=0.95)

print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})")
print(f"Margin of Error: {moe:.2f}")
```

Q10.write a python program to calculate the margin of error for a given confidence level using sample data.

```python
import numpy as np
from scipy import stats

def calculate_margin_of_error(sample_data, confidence_level=0.95):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation

    # Calculate the critical t-value
    t_critical = stats.t.ppf((1 + confidence_level) / 2, df=n - 1)

    # Standard Error of the Mean (SEM)
    standard_error = sample_std / np.sqrt(n)

    # Margin of Error
    margin_of_error = t_critical * standard_error

    return margin_of_error, sample_mean, t_critical, standard_error

# Example usage
np.random.seed(42)
data = np.random.normal(loc=50, scale=12, size=25)
moe, mean, t_val, sem = calculate_margin_of_error(data, confidence_level=0.95)

print(f"Sample Mean: {mean:.2f}")
print(f"Standard Error (SE): {sem:.4f}")
print(f"T-critical value: {t_val:.4f}")
print(f"Margin of Error (95% CI): Â±{moe:.2f}")
```

Q11.implement a bayesian inference method using bayes' Theorem in python and explain the process?

```python
def bayesian_inference(prior, likelihood, false_positive_rate):
    # Prior: P(Disease) = prior
    # Likelihood: P(Positive | Disease)
    # False Positive: P(Positive | No Disease) = false_positive_rate
    # P(No Disease) = 1 - prior

    # Total probability of testing positive
    evidence = likelihood * prior + false_positive_rate * (1 - prior)

    # Posterior probability using Bayes' Theorem
    posterior = (likelihood * prior) / evidence

    return posterior, evidence

# Example values
prior = 0.01                     # 1% have the disease
likelihood = 0.99               # True positive rate
false_positive_rate = 0.05      # False positive rate

posterior, evidence = bayesian_inference(prior, likelihood, false_positive_rate)

print(f"Evidence (P(Positive)): {evidence:.4f}")
print(f"Posterior (P(Disease | Positive)): {posterior:.4f}")
```
